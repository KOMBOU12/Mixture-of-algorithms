```python
import pandas as pd
import matplotlib.pyplot as plt
college = pd.read_csv('College.csv')
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'},
axis=1)
college3 = college3.set_index('College')
college = college3

#Réponse à la question 8.c
print(college.describe())

# Réponse à la question 8.d
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']], alpha=0.2, figsize=(8, 8), diagonal='hist')
plt.show


```

```python
#Réponse à la question 8.e
#college.boxplot(column='Outstate')
#college.boxplot(column='Personal')
# Create side-by-side boxplots for 'Outstate' and 'Personal'
college.boxplot(column=['Outstate', 'Personal'])
# Add title and labels
plt.title('Side-by-Side Boxplots')

plt.ylabel('Outstate  Personal')

# Show the plot
plt.show()
```

```python
#Réponse à la question 8.f
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
print(college[['Elite']])
college.boxplot(column=['Outstate'], by='Elite')
# Add title and labels
plt.title('Side-by-Side Boxplots')
plt.xlabel('Elite')
plt.ylabel('Outstate')

# Show the plot
plt.show()
```

```python
#Réponse à la question 8.g
# Sélectionnez la colonne que vous souhaitez représenter sous forme d'histogramme
column_to_plot = 'Top10perc'

# Créez l'histogramme en utilisant Matplotlib
plt.hist(college[column_to_plot], bins=10, edgecolor='k', alpha=0.7)
plt.title(f'Histogramme de {column_to_plot}')
plt.xlabel(column_to_plot)
plt.ylabel('Fréquence')

# Affichez l'histogramme
plt.show()

```
